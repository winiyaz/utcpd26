# 238 - Filter Even Numbers 

```python
# Lets create a random list of random even mumbers 
import random as rp
ra1e = [x for x in range(0, 31, 2)]
rprint(ra1e)
```

```python
# Create a random list of numbers adn then filter by even numbers
li1 = [n for n in range(0,21)]
rprint(li1)
```

```python
# Now from li1 do filter fro even numbers
li1_even = [n for n in li1 if n%2==0]
rprint(f"[green] Even Numbers")
rprint(li1_even)

#Now from li1 do filter for odd numbers
li1_odd = [n for n in li1 if n%2!=0]
rprint(f"[blue] Odd Numbers")
rprint(li1_odd)
```

# Dictionary Comprehension

1. Used for creating a  dictionary 


```python
# Creating a list of names 
stu = [
    "Scarlett",
    "Sofia",
    "Ruby",
    "Aurora",
    "Luna",
    "Layla",
    "Eva",
    "Sienna",
    "Daisy",
    "Zoe",
    "Ivy",
    "Freya",
    "Jade",
    "Charlotte",
    "Sophie",
    "Ember",
    "Alexandra",
    "Georgia",
    "Ophelia",
    "Valentina"
]
```

1. Below now using dictionary comprehension

```python
import random as rp
stu_sco = {s:rp.randint(1,100) for s in stu}
rprint(stu_sco)
```

1. Code below using the dictionary and and then making a new dicttionary where the value in the key:value pair is more than a specifc number 
    1. 

```python
# using stu_sco 

pass_stu = {stu:sco for (stu, sco) in stu_sco.items() if sco >= 60}
rprint(pass_stu)
```
